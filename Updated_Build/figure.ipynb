```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cold_1m = np.load('cold_training_1.0m_21_False_0.2_0.01_1.5.npy')
hot_1m = np.load('hot_training_1.0m_21_False_0.2_0.01_1.5.npy')

cold_3m = np.load('cold_training_3.0m_65_False_0.2_0.01_1.5.npy')
hot_3m = np.load('hot_training_3.0m_65_False_0.2_0.01_1.5.npy')
```

```python
# i want two subplots, one for 1m and one for 3m
fig, axs = plt.subplots(1, 2, figsize=(18, 6))
# plot the first subplot
axs[0].plot(cold_1m, label='Cold Start', color='blue')
axs[0].plot(hot_1m, label='Hot Start', color='red')
axs[0].axhline(y=-350, color='k', linestyle='--', linewidth=1.5, label = 'Black-Scholes')
axs[0].axhline(y=-520, color='k', linestyle='-', linewidth=1.5, label = 'No Hedging')
axs[0].set_xlabel('Episodes')
axs[0].set_ylabel('Reward')
axs[0].legend(loc = 'center right')
axs[0].set_ylim(-3000, 0)
axs[0].set_xlim(0, 50000)
axs[0].grid()
#axs[0].set_xticks(np.arange(0, 101, 10))
#axs[0].set_xticklabels(np.arange(0, 101, 10), rotation=45)
#axs[0].set_yticks(np.arange(0, 1.1, 0.1))
#axs[0].set_yticklabels(np.arange(0, 1.1, 0.1), rotation=45) 
axs[0].set_title('Training curve for 1-month expiry option')
# plot the second subplot
axs[1].plot(cold_3m, label='Cold Start', color='blue')
axs[1].plot(hot_3m, label='Hot Start', color='red')
axs[1].axhline(y=-498, color='k', linestyle='--', linewidth=1.5, label = 'Black-Scholes')
axs[1].axhline(y=-913, color='k', linestyle='-', linewidth=1.5, label = 'No Hedging')
axs[1].set_xlabel('Episodes')
axs[1].set_ylabel('Reward')
axs[1].legend(loc = 'center right')
axs[1].set_ylim(-5000, 0)
axs[1].set_xlim(0, 50000)
axs[1].grid()
#axs[1].set_xticks(np.arange(0, 101, 10))
#axs[1].set_xticklabels(np.arange(0, 101, 10), rotation=45)
#axs[1].set_yticks(np.arange(0, 1.1, 0.1))
#axs[1].set_yticklabels(np.arange(0, 1.1, 0.1), rotation=45)
axs[1].set_title('Training curve for 3-month expiry option')
# set the title for the whole figure
#fig.suptitle('Comparing training curves for ', fontsize=16)

```

```python
plt.figure(figsize=(10, 6))
plt.plot(cold_1m, label = 'Cold Training', color = 'blue')
plt.plot(hot_1m, label = 'Hot Training', color = 'red')
plt.title('Training Loss')
plt.xlabel('Episodes')
plt.ylabel('Loss')
plt.legend(loc = 'center right')
```
